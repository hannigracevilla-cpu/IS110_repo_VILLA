**Exercise 2**

Enter Hours: 45

Enter Rate: 10

Pay: 475.0

```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    overtime = hours - 40
    pay = (40 * rate) + (overtime * rate * 1.5)
else:
    pay = hours * rate

print("Pay:", pay)
```

```python
score = float(input("Enter score: "))

if score >= 98 and score <= 100:
    print("Grade: 1.0")
elif score >= 94 and score <= 97:
    print("Grade: 1.25")
elif score >= 90 and score <= 93:
    print("Grade: 1.5")
elif score >= 86 and score <= 89:
    print("Grade: 1.75")
elif score >= 82 and score <= 85:
    print("Grade: 2.0")
elif score >= 78 and score <= 81:
    print("Grade: 2.25")
elif score >= 75 and score <= 77:
    print("Grade: 2.5")
elif score >= 0 and score <= 74:
    print("Grade: 5.0")
else:
    print("Invalid score")
```
