# `Exercise A`

```python
import numpy as np  # Import the NumPy library for numerical operations

# Create a nested Python list with values from 1 to 9
data = [[1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]]

A = np.array(data)  # Convert the nested list into a NumPy ndarray

print(A)  # Display the array
```

# Using `ndarray property, method of the ndarray object, using numpy function and passing an ndarray`

```python
import numpy as np  # Import NumPy library

data = [[10, 20, 30],
        [40, 50, 60],
        [70, 80, 90]]

A = np.array(data)  # Convert list to ndarray

print("Shape:", A.shape)  # Get the dimensions of the array

print("max:", A.max())  # Get the maximum value in the array

print("max(A):",np.max(A))  # Get the maximum value using NumPy
```

# `Explanation `
<a id='Explanation '></a>

- A nested list was created and converted into a NumPy ndarray.
- A.shape was used to check the dimensions of the array.
- A.max() returns the highest value using an ndarray method.
- np.max(A) returns the same result using a NumPy function.

# `Exercise B`
<a id='Exercise B'></a>


```python
import numpy as np  # Import NumPy library

B = np.ones((5, 5), dtype=int)  # Create a 5x5 array filled with 1s (integer type)

print(B)  # Display the array

print("Shape:", B.shape)  # Check the shape of the array
print("Size:", B.size)   # Check the total number of elements
print("NDim:", B.ndim)   # Check the number of dimensions
```

# `Explanation`
<a id='Explanation'></a>

- np.ones() was used to create a 5x5 array filled with ones.
- B.shape shows the dimensions (5x5).
- B.size shows the total number of elements (25).
- B.ndim shows that the array is 2-dimensional.

# `Exercise C`
<a id='Exercise C'></a>

```python
import numpy as np  # Import NumPy library

C = np.arange(-5, 6)  # Create an array from -5 to +5

result = C * 2  # Multiply each element by 2

print(result)  # Display the resulting array
```

# `Explanation`
<a id='Explanation'></a>

- np.arange(-5, 6) generates values from -5 to 5.
- Each value was multiplied by 2 using element-wise operation.
- The result shows the transformed array.
